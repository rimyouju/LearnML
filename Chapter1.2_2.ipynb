## 코랩과 주피터 노트북

### 구글 코랩

- 웹 브라우저에서 무료로 파이썬 프로그램을 테스트하고 저장할 수 있는 서비스

- 머신러닝 프로그램도 만들 수 있음
- 클라우드 기반의 주피터 노트북 개발 환경
- 구글 코랩을 사용하면 컴퓨터 성능과 상관없이 프로그램을 실습 가능

- 노트북(코랩 노트북)에서 셀(cell)은 코드 또는 텍스트의 덩어리

- 노트북은 보통 여러 개의 코드 셀과 텍스트 셀로 이루어짐
- 키보드 위아래 키를 누르면 셀 사이를 이동 가능

### 텍스트 셀

셀(cell)
- 코랩에서 실행할 수 있는 최소 단위

- 셀 안에 있는 내용을 한 번에 실행하고 그 결과를 노트북에 나타냄
- 텍스트 셀은 코드처럼 실행되는 것은 아니기 때문에 자유롭게 사용 가능
- 셀 하나에 아주 긴 글을 써도 되고 여러 셀에 나누어 작성도 가능

- 노트북을 사용하면 코드를 설명하는 문서를 따로 만들지 않고 코드와 텍스트를 함께 담을 수 있음

- 코드의 실행 결과도 노트북과 함께 저장
- 다른 사람에게서 노트북 파일을 받으면 코드를 실행할 필요 없이 코드 설명과 실행 결과를 바로 확인 가능

### 노트북

- 코랩은 구글이 대화식 프로그래밍 환경인 주피터(jupyter)를 커스터마이징한 것

- 파이썬 지원으로 시작한 주피터 프로젝트는 최근에는 다른 언어도 지원
- 주피터 프로젝트의 대표 제품이 주피터 노트북(notebook)

- 코랩 노트북은 구글 클라우드의 가상 서버(virtual machine)를 사용

- 코랩 노트북으로 동시에 사용할 수 있는 구글 클라우드의 가상 서버는 최대 5개

- 5개 이상의 노트북을 열어야 한다면 이미 실행 중인 노트북을 저장한 다음 구글 클라우드와 연결을 끊어야 함
