### 인공지능이란

인공지능

- 인공지능 황금기 -> 프랑크 로젠블라트 - 퍼셉트론           
(첫 번째 AI 겨울; 컴퓨터 성능의 한계로 인해 간단한 문제를 해결하는 것에 그침)
- 전문가 시스쳄 등장(두 번째 AI 붐)           
(두 번째 AI 겨울)            
- 인공일반지능(artificiaal general intelligence) 또는 강인공지능(strong AI)

현실에서 우리가 마주하고 있는 인공지능 -> 약인공지능(Week AI)

### 머신러닝이란

머신러닝              
: 규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야;   
인공지능의 하위 분야 중에서 지능을 구현하기 위한 소프트웨어를 담당하는 핵심 분야
- 머신러닝은 통계학과 깊은 관련이 있음
- 최근 머신러닝의 발전은 통계나 수학 이론보다 경험을 바탕으로 발전하는 경우도 많음

사이킷런(scikit-learn) -> 컴퓨터 과학 분야의 대표적인 머신러닝 라이브러리        
- 파이썬 API를 사용 -> 배우기 쉽고 컴파일하지 않아도 되기 때문에 사용하기 편리

- 머신러닝 라이브러리에 포함된 알고리즘들은 안정적이며 성능이 검증됨
- 누구나 머신러닝 알고리즘을 무료로 손쉽게 제품에 활용 가능

## 마켓과 머신러닝

- 보통의 프로그램 -> 누군가 정해준 기준대로 일

- 머신러닝 -> 스스로 기준을 찾아서 일

이진 분류
- 2개의 클래스 중 하나를 고르는 문제

- 분류 : 머신러닝에서 여러 개의 종류(혹은 클래스) 중 하나를 구별해 내는 문제

```python
# BreamAndSmelt
# 도미 데이터 준비하기

# 도미의 길이
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미의 무게
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

- 특성 -> 데이터의 특징
-> 각 도미의 길이와 무게

- 산점도 : x, y축으로 이루어진 좌표계에 두 변수(x, y)의 관계를 표현하는 방법

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

- 산점도 그래프가 일직선에 가까운 형태로 나타나는 경우 -> 선형적이다

```python
# 빙어 데이터 준비하기

# 빙어의 길이
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

# 빙어의 무게
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# 빙어 그래프 그리기
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 첫 번째 머신러닝 프로그램

- k-최근접 이웃(k-Nearest Neighbors) 알고리즘 사용

```python
# 첫 번째 머신러닝 프로그램

# 리스트 형태 -> [도미1, 도미2,.., 도미n, 빙어1, 빙어2,..., 빙어n]
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

- 사이킷런(scikit-learn) 패키지 사용

- 각 특성의 리스트를 세로 방향으로 늘어뜨린 2차원 리스트를 만들어야 함

- zip() 함수와 내포(list comprehension) 구문 사용

- zip() 함수 : 나열된 리스트 각각에서 하나씩 원소를 꺼내 반환

```python
# 2차원 리스트 생성
fish_data = [[l, w] for l, w in zip(length, weight)]
print(fish_data)
```

```python
# 컴퓨터 프로그램이 이해할 수 있게 도미와 방어를 1과 0으로 표현
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 클래스 객체 생성
kn = KNeighborsClassifier()
```

```python
# 객체에 fish_data와 fish_target을 전달하여 기준 학습(훈련)
# 훈련 -> 모델에 데이터를 전달하여 규칙을 학습하는 과정

# fit() 메서드 활용 -> 주어진 데이터로 알고리즘 활용
kn.fit(fish_data, fish_target)
```

```python
# 훈련 평가(정확도 구하기)
# score 메서드 활용 -> 0에서 1 사이의 값 반환
kn.score(fish_data, fish_target)
```

k-최근접 이웃 알고리즘
- 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 사용

- 주위의 데이터로 현재 데이터를 판단

- 사실상 어떤 규칙을 찾기보다는 전체 데이터를 메모리에 가지고 있는 것이 전부

```python
# k-최근접 이웃 알고리즘
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 200, marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30, 600]])
```

- predict() 메서드 -> 새로운 데이터의 정답을 예측; 2차원 리스트를 전달해야 함

- k-최근접 이웃 알고리즘을 위해 준비해야 할 일 -> 데이터를 모두 가지고 있는 것

- 새로운 데이터에 대해 예측할 때는 가장 가까운 직선거리에 어떤 데이터가 있는지를 살피면 됨
- k-최근접 이웃 알고리즘은 데이터가 아주 많은 경우 사용하기 어려움
- 데이터가 크기 때문에 메모리가 많이 필요
- 계산 시간 많이 소요됨

- KNeighborsClassifier 클래스의 _fit_X 속성에 fish_data를 모두 가지고 있음

- _y 속성에 fish_target을 가지고 있음

```python
print(kn._fit_X)
```

```python
print(kn._y)
```

```python
# KNeighborsClassfier 클래스의 기본값은 5
kn49 = KNeighborsClassifier(n_neighbors = 49) # 참고 데이터를 49개로 한 kn49 모델
```

```python
# fish_data의 49개 중에 도미가 35개를 차지하므로 어떤 데이터를 넣어도 무조건 도미로 예측
kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
# 정확도 계산
# 정확도 -> 정확한 답을 몇 개 밎혔는지를 백분율로 나타낸 값
# 정확도 = (정확히 맞힌 개수) / (전체 데이터 개수)
print(35/49)
```
